```python
#Exploratory data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Iris.csv')
df.shape
```

```python
df.head()
```

```python
df = df.drop(['Id'],axis=1)
df.head()
```

```python
df['Species'].value_counts()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.plot()
plt.show()
```

```python
df.plot(kind='scatter',x='SepalLengthCm',y='PetalLengthCm')
plt.show()
```

```python
sns.FacetGrid(df, hue='Species',height = 6).map(plt.scatter,'SepalLengthCm','PetalLengthCm').add_legend()
plt.show()
```

```python
sns.pairplot(df,hue='Species')
plt.show()
```

```python
sns.FacetGrid(df, hue='Species',height = 6).map(plt.scatter,'PetalLengthCm','PetalWidthCm').add_legend()
plt.show()
```

```python
sns.FacetGrid(df, hue='Species',height = 6).map(sns.histplot,'SepalLengthCm').add_legend()
plt.show()
```

```python
sns.FacetGrid(df, hue='Species',height = 6).map(sns.histplot,'PetalLengthCm').add_legend()
plt.show()
```
